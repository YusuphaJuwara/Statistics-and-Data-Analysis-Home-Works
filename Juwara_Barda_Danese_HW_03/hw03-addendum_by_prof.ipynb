## Graph in Python with [`igraph`](https://igraph.org/python/)

```python
# url: https://igraph.org/python/
# pip install igraph
import igraph as ig
print(ig.__version__)
```

To create a graph from an adjecency matrix, use 
`Graph.Adjacency()` or, for weighted matrices, `Graph.Weighted_Adjacency()`:

```python
# Directed (by default)
# url: https://igraph.org/python/tutorial/latest/generation.html#from-matrices
g_dir = ig.Graph.Adjacency([[0, 1, 1], [0, 0, 0], [0, 0, 1]])
g_dir.vs["name"] = ["Pampulo", "Pimpulo", "Franco"]  # add additional info to the vs (vertex set)
print(g_dir)
g_dir.get_adjacency()
```

```python
# Undirected
g_undir = ig.Graph.Adjacency([[0, 1, 1], [0, 0, 0], [0, 0, 1]], mode = 'undirected')
g_undir.vs["name"] = ["Pampulo", "Pimpulo", "Franco"]  # add additional info to the vs (vertex set)
print(g_undir)
g_undir.get_adjacency()
```

### Visualization

To visualize the graph, notice that plotting is dependent on the 
[pycairo](http://www.cairographics.org/pycairo) or [cairocffi](http://cairocffi.readthedocs.io/) libraries that provide Python bindings to the popular 
[Cairo library](http://www.cairographics.org/). 

```python
# url: https://igraph.org/python/tutorial/latest/visualisation.html
# https://igraph.org/python/api/0.9.8/igraph.drawing.html
# pip install cairocffi
ig.plot(g_dir, layout="kamada_kawai")
```

```python
# url: https://igraph.org/python/tutorial/latest/tutorial.html#vertex-attributes-controlling-graph-plots
g_undir.vs["label"] = g_undir.vs["name"]
ig.plot(g_undir, layout="random",
       bbox=(200, 200), margin=30)
```
